# Analysis of some outliers in the output OneError

We can see in the video that the algorithm works pretty well except at minute 24 and minute 40 where there are some outliers



<img src="../output_video/OneError/A1.jpg">

To analyze this we have a csv with the data and we plot it

```python
headers=['LA','LB','LC','RA','RB','RC']
df = pd.read_csv("../output_video/OneError/log.csv",names=headers)
```

```python
df.head()
```

```python
Y1=df['LA']
Y2=df['RA']

# Plot the result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.plot(Y1)
ax1.set_title('Left Quadratic Coef', fontsize=50)
ax2.plot(Y2)
ax2.set_title('Right Quadratic Coef', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

```python
plt.figure(figsize=(20,10))
plt.plot(Y1)
plt.plot(Y2)
```
